# Job Posting Collection [Script E]

*This notebook contains the script used to collect job postings from Glassdoor. This scraper was built upon an open-source Glassdoor job postings scraper posted on Github. This script makes a call to the open-source API for each company in our company list, and collects at least 100 job postings for them. The final output of this script is the Glassdoor Job Postings dataset.*

---
*References: https://github.com/kelvinxuande/glassdoor-scraper*


## Package Installation, Imports & Setup

```python
import pandas as pd
import json
import os
```

## Data Import

```python
companies = pd.read_csv('company_list.csv')
company_list = companies.to_dict('records')
```

```python
for company in company_list:
    url = 'https://www.glassdoor.sg/Job/us-data-jobs-SRCH_IL.0,2_IN1_KO3,7.htm?companyId='+str(company['company_id'])[1:]
    config = {"base_url":url, "target_num":20, "company_name":company['company_name']}
    with open('glassdoor-scraper-master/src/data/config.json', 'w') as fp:
        json.dump(config, fp)
    os.system('python glassdoor-scraper-master/src/main.py')
```
